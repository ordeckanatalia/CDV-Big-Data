```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
tur_2017 = pd.read_csv('tur_2017.csv', delimiter =';')
tur_2018 = pd.read_csv('tur_2018.csv', delimiter =';')
tur_2019 = pd.read_csv('tur_2019.csv', delimiter =';')
tur_2020 = pd.read_csv('tur_2020.csv', delimiter =';')
```

```python
def clean_data(name):
    name = name.iloc[:, 1 : 13]
    name.columns = ['Nazwa', 'styczeń', 'luty', 'marzec', 'kwiecien', 'maj', 'czerwiec', 'lipiec', 'sierpien', 'październik', 'listopad', 'grudzień']
    name = name.reset_index().drop(columns=['index'])
    name = name.set_index('Nazwa')
    name = name.transpose()
    return name
```

```python
tur_2017 = clean_data(tur_2017)
tur_2018 = clean_data(tur_2018)
tur_2019 = clean_data(tur_2019)
tur_2020 = clean_data(tur_2020)
tur_2017.head()
```

```python
tur_2017['POLSKA'].plot(kind = 'bar', color='blue')
```

```python
tur_2018['POLSKA'].plot(kind = 'bar', color='#DC5039')
```

```python
tur_2019['POLSKA'].plot(kind = 'bar', color='#FBA40A')
```

Jak widać na poprzednich wykresach, ilość turystów w Polsce, po krótkiej zimowej stagnacji, w okolicach marca zaczynała powoli rosnąć. Był to trend, który utrzymywał się przez kilka poprzednich lat. Jednak w tym roku liczba podróżujących w marcu drastycznie zmalała, na co wpływ miały ograniczenia wytyczone przez rząd i obawy przed chorobą.

```python
tur_2020['POLSKA'].plot(kind = 'bar', color='#410967')
```

```python
tur_2017_PL = tur_2017.iloc[:, 0 :1]
tur_2018_PL = tur_2018.iloc[:, 0 :1]
tur_2019_PL = tur_2019.iloc[:, 0 :1]
tur_2020_PL = tur_2020.iloc[:, 0 :1]
```

```python
tur_2017_PL.columns = ['2017']
tur_2018_PL.columns = ['2018']
tur_2019_PL.columns = ['2019']
tur_2020_PL.columns = ['2020']
```

```python
tur_all_PL = pd.merge(tur_2017_PL, tur_2018_PL, right_index=True, left_index=True)
tur_all_PL = pd.merge(tur_all_PL, tur_2019_PL, right_index=True, left_index=True)
tur_all_PL = pd.merge(tur_all_PL, tur_2020_PL, right_index=True, left_index=True)
```

Poniższy wykres pokazuje, że ilość turystów w poprzednich latach systematycznie się zwiększała. Tendencja wzrostowa załamała się jednak na początku marca z powodu wybuchu pandemii koronawirusa.

```python
plt.figure(figsize = (16,5))
plt.plot(tur_all_PL['2017'], color='blue',label='2017')
plt.plot(tur_all_PL['2018'], color='#DC5039',label='2018')
plt.plot(tur_all_PL['2019'],color='#FBA40A',label='2019')
plt.plot(tur_all_PL['2020'],color='#410967',label='2020')
plt.legend(loc='best')
plt.title('Ilość turystów w Polsce w kolejnych latach')
```

```python
zaw = pd.read_csv('zawieszeni.csv', delimiter =';',  encoding = "ISO-8859-1")
zaw = zaw.iloc[:, 0 : 3]
zaw.columns = ['rok', 'wojewodztwo', 'zawieszeni przedsiebiorcy']
zaw.head()
```

```python
razem_zaw = zaw[zaw['wojewodztwo'] == 'lacznie']
```

Jak widać na poniższym wykresie, 2020 rok nie był łaskawy dla przediębiorców turystycznych - działalność zawiesiło ponad 3 razy więcej przedsiębiorstw niż mialo to miejsce w poprzednim roku, a prezentowane dane przedstawiają stan z 25.05.2020.

```python
razem_zaw.plot(x= 'rok', y= 'zawieszeni przedsiebiorcy', kind = 'bar', color='#DC5039')
```

```python
woj_zaw_2020 = zaw[zaw['rok'] == 2020]
woj_zaw_2020 = woj_zaw_2020[woj_zaw_2020['wojewodztwo'] != 'lacznie']
```

Poniższy wykres pokazuje, że najwięcej przedsiębiorców w 2020 roku zrezygnowało z działalności w województwie mazowieckim, a najmniej w lubuskim. Wiele przedsiębiorców zrezygnowało też z dzialalności na terenach atrakcyjnych turystycznie, takich jak województwo małopolskie (Kraków, Zakopane), czy województwo pomorskie (morze).

```python
woj_zaw_2020.plot(x= 'wojewodztwo', y= 'zawieszeni przedsiebiorcy', kind = 'bar', color='#FBA40A')
```

Jak widać na powyższych wykresach, światowa pandemia w znacznym stopniu wpłynęła na przedsiębiorstwa turystyczne w Polsce. Może to być znak, że niektórzy zrezygnują z prowadzenia swoich działalności, co zmniejszy konkurencję i łatwiej będzie wejść na rynek nowym przedsiębiorcom kiedy pandemia dobiegnie końca.
