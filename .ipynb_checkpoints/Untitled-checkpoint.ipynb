Importuję niezbędne pakiety:

```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import geopandas as gpd

%matplotlib inline
```

Zapisuję moje dane do zmiennej:

```python
dane_turystyka_ogolny = 'turystyka1.csv'
```

Wczytuję i przeglądam dane:

```python
noclegi_ogolny = pd.read_csv(dane_turystyka_ogolny, delimiter =';')
noclegi_ogolny.head()
```

Widzę, że moje dane mają niepoprawne nazwy kolumn i jedną dodatkową kolumnę na końcu. Problemem mogą być też nieistniejące wartości, z którymi trzeba będzie coś zrobić

```python
noclegi_ogolny = noclegi_ogolny.iloc[:, 0 : 12]
noclegi_ogolny.columns = ['TERYT', 'Nazwa', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019']
noclegi_ogolny.head()
```

```python
noclegi_ogolny.tail()
```

```python
noclegi_ogolny.info()
```

```python
# kod ma być stringiem o stałejdługości (dodanie wiodących zer)
noclegi_ogolny['TERYT_gmn'] = noclegi_ogolny.TERYT.apply(lambda x: '0'+str(x) if len(str(x)) < 7 else str(x))
noclegi_ogolny.head()
```

```python
#robię też tabelę tylko dla województw, bo one przydadzą mi się do analizy
noclegi_ogolny_woj = noclegi_ogolny[noclegi_ogolny.TERYT_gmn.str[2:7] == '00000']
noclegi_ogolny_woj
```

```python
#tu zrobię wykresy
noclegi_ogolny_woj.plot(x = 'Nazwa', y = ['2018', '2019'])
```
